#Theoretical

#1. What is a Support Vector Machine (SVM)

Ans:- A Support Vector Machine (SVM) is a machine learning model used for classification and regression tasks. It finds the best boundary (hyperplane) to separate data points into different classes. SVM works well with both linear and non-linear data using kernels.

#2. What is the difference between Hard Margin and Soft Margin SVM

Ans:- Hard Margin SVM: It finds a perfect separating boundary but works only if data is completely separable.

Soft Margin SVM: It allows some misclassification to handle overlapping data and improves generalization.

Soft Margin is more flexible and works better in real-world cases

#3.  What is the mathematical intuition behind SVM

Ans:- The main idea behind SVM is to find a hyperplane that maximizes the margin between two classes. The margin is the distance between the hyperplane and the nearest data points from each class (called support vectors). The goal is to maximize this margin while ensuring correct classification. This can be formulated as an optimization problem.

#4. What is the role of Lagrange Multipliers in SVM

Ans:- Lagrange Multipliers in SVM help to optimize the objective function while satisfying the constraints (correctly classifying data). They allow us to convert the constrained optimization problem into an unconstrained one, making it easier to solve. The Lagrange multipliers are associated with the support vectors, indicating their importance in defining the optimal hyperplane.

#5. What are Support Vectors in SVM

Ans:- Support Vectors are the data points that are closest to the decision boundary (hyperplane) in an SVM. These points are crucial because they determine the optimal hyperplane that separates the classes. The SVM uses only these points to define the boundary, making them the most influential for the model's performance.

#6. What is a Support Vector Classifier (SVC)

Ans:- A Support Vector Classifier (SVC) is an SVM model used for classification tasks. It finds the optimal hyperplane that best separates different classes in the dataset, using support vectors to define the boundary. SVC can handle both linear and non-linear data by using different kernels.

#7. What is a Support Vector Regressor (SVR)

Ans:- A Support Vector Regressor (SVR) is an SVM model used for regression tasks. It aims to find a hyperplane that best fits the data, while allowing some errors within a certain margin. SVR tries to make predictions while keeping the margin as wide as possible for better generalization.

#8. What is the Kernel Trick in SVM

Ans:- The Kernel Trick in SVM is a technique that transforms non-linear data into a higher-dimensional space, making it easier to find a linear separation. It uses a mathematical function (called a kernel) to compute the dot product of data points in this new space without explicitly transforming them, saving computation time.

#9.  Compare Linear Kernel, Polynomial Kernel, and RBF Kernel

Ans:- Linear Kernel: It is used when the data is linearly separable. It tries to find a straight line (or hyperplane) to separate the classes.

Polynomial Kernel: It transforms data into a higher-dimensional space using polynomial functions. It's useful for non-linear data but with a defined relationship between features.

RBF (Radial Basis Function) Kernel: It is the most flexible and can handle complex non-linear data. It calculates the similarity between points using a Gaussian function and can work well even with no clear separation between classes.

#10. What is the effect of the C parameter in SVM

Ans:- The C parameter in SVM controls the trade-off between maximizing the margin and minimizing classification errors.

High C: Gives more importance to correctly classifying training data, resulting in a smaller margin but fewer misclassifications.

Low C: Allows more misclassifications in exchange for a larger margin, improving generalization to new data.

#11. What is the role of the Gamma parameter in RBF Kernel SVM

Ans:- The Gamma parameter (Œ≥) in RBF Kernel SVM controls how far the influence of a single data point reaches.

High Gamma: The model focuses on nearby points, leading to high variance (overfitting).

Low Gamma: The model considers distant points, making it more generalized but possibly underfitting.

It determines how complex the decision boundary will be.

#12. What is the Na√Øve Bayes classifier, and why is it called "Na√Øve"?

Ans:- The Na√Øve Bayes classifier is a simple but powerful probabilistic model based on Bayes' Theorem. It is mainly used for classification tasks like spam filtering and sentiment analysis.

It is called "Na√Øve" because it assumes that all features are independent, which is rarely true in real-world data, but this assumption makes the model fast and efficient.

#14. Explain the differences between Gaussian Na√Øve Bayes, Multinomial Na√Øve Bayes, and Bernoulli Na√Øve Bayes.

Ans:- Differences Between Na√Øve Bayes Variants
Gaussian Na√Øve Bayes

Used for continuous data that follows a normal (Gaussian) distribution.

Example: Classifying iris flowers based on petal length/width.

Multinomial Na√Øve Bayes

Used for discrete data, like word counts in text classification.

Example: Spam detection in emails (based on word frequency).

Bernoulli Na√Øve Bayes

Used for binary data (0s and 1s), where features are either present or absent.

Example: Sentiment analysis (word exists in a sentence or not).

Each type is optimized for different kinds of data.

#15. When should you use Gaussian Na√Øve Bayes over other variants?

Ans:- You should use Gaussian Na√Øve Bayes when your features are continuous and follow a normal (Gaussian) distribution.

‚úÖ Best for:

Iris classification (petal length/width)

Medical diagnoses (blood pressure, cholesterol levels)

Financial data (stock prices, risk scores)

If your data is text-based (word counts), use Multinomial Na√Øve Bayes, and for binary (0/1) data, use Bernoulli Na√Øve Bayes.

#16.  What are the key assumptions made by Na√Øve Bayes

Ans:- The Na√Øve Bayes classifier makes two key assumptions:

Feature Independence üõë

It assumes that all features are independent of each other, meaning the presence of one feature does not affect another. (This is rarely true in real-world data!)

Equal Importance of Features ‚öñÔ∏è

It assumes all features contribute equally to the final classification, even if some are more important than others.

Despite these "na√Øve" assumptions, Na√Øve Bayes works well in many practical applications.

#24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

X, y = make_classification(n_samples=100, n_features=2, n_informative=2,
                           n_redundant=0, n_repeated=0, n_clusters_per_class=1,
                           random_state=1)

svm_poly = make_pipeline(StandardScaler(), SVC(kernel='poly', degree=3, C=1.0))

svm_poly.fit(X, y)

def plot_decision_boundary(model, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('SVM with Polynomial Kernel')
    plt.show()

plot_decision_boundary(svm_poly, X, y)
```

#38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy

```python
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, cross_val_score

# Load Iris dataset
iris = datasets.load_iris()
X, y = iris.data, iris.target

# Define Stratified K-Fold cross-validation
k = 5  # Number of folds
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=1)

# Define SVM model
svm_model = SVC(kernel='linear')

# Perform Cross-Validation
scores = cross_val_score(svm_model, X, y, cv=skf, scoring='accuracy')

# Print accuracy for each fold
for i, score in enumerate(scores):
    print(f"Fold {i+1}: Accuracy = {score:.4f}")

# Compute and print average accuracy
average_accuracy = np.mean(scores)
print(f"\nAverage Accuracy: {average_accuracy:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve

# Load Breast Cancer dataset
data = load_breast_cancer()
X, y = data.data, data.target  # Features and labels

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train a Na√Øve Bayes Classifier
nb = GaussianNB()
nb.fit(X_train, y_train)

# Get probability predictions for ROC curve
y_probs = nb.predict_proba(X_test)[:, 1]  # Probability of class 1

# Compute ROC-AUC Score
roc_auc = roc_auc_score(y_test, y_probs)
print(f"ROC-AUC Score: {roc_auc:.4f}")

# Compute ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_probs)

# Plot ROC Curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='blue', label=f'Na√Øve Bayes (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random chance line
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve for Na√Øve Bayes Classifier")
plt.legend()
plt.show()
```

#46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import precision_recall_curve, auc

# Load the Breast Cancer dataset
data = load_breast_cancer()
X, y = data.data, data.target  # Features and labels

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Train an SVM Classifier with probability estimates enabled
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)

# Get probability scores for PR curve
y_probs = svm.predict_proba(X_test)[:, 1]  # Probability of class 1

# Compute Precision-Recall curve
precision, recall, _ = precision_recall_curve(y_test, y_probs)

# Compute Area Under the Curve (AUC) for PR curve
pr_auc = auc(recall, precision)

# Plot Precision-Recall Curve
plt.figure(figsize=(6, 4))
plt.plot(recall, precision, color='blue', label=f'SVM (AUC = {pr_auc:.4f})')
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve for SVM Classifier")
plt.legend()
plt.grid()
plt.show()
```
