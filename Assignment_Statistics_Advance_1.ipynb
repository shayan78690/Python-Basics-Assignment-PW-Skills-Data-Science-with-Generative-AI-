#Theory Questions

#1. What is a random variable in probability theory?

A random variable is a numerical outcome of a random phenomenon. It can be discrete (finite/countable values) or continuous (infinite possible values).

#2. What are the types of random variables?

There are two types: Discrete random variables (take distinct values, e.g., dice rolls) and Continuous random variables (take any value within a range, e.g., height, weight).





#3.What is the difference between discrete and continuous distributions?

Discrete distributions describe variables that take specific values (e.g., number of heads in coin tosses), while continuous distributions describe variables over a continuous range (e.g., height of people).



#4. What are probability distribution functions (PDF)?

A PDF describes the likelihood of a random variable taking a particular value. For discrete variables, it's called a probability mass function (PMF).



#5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

A PDF (or PMF) gives the probability of a specific value, while a CDF gives the probability that a variable takes a value less than or equal to a given number.

#6. What is a discrete uniform distribution?

A distribution where all possible outcomes have equal probability, e.g., rolling a fair die (each number from 1 to 6 has a 1/6 chance).

#13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

CLT states that the distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population's original distribution.

#14. How does the Central Limit Theorem relate to the normal distribution?

CLT ensures that even if the original data is not normally distributed, the sampling distribution of the mean will be approximately normal for large enough samples.

#15. What is the application of Z statistics in hypothesis testing?

Z-statistics help determine how far a sample mean is from a population mean in terms of standard deviations, enabling hypothesis testing and confidence intervals.

#17. What are point estimates and interval estimates in statistics?

A point estimate is a single value approximation of a parameter (e.g., sample mean). An interval estimate provides a range (e.g., confidence interval) where the parameter is likely to lie.

#18. What is the significance of confidence intervals in statistical analysis?

A confidence interval provides a range of values within which the true population parameter is expected to fall, with a given level of confidence (e.g., 95%).

#19. What is the relationship between a Z-score and a confidence interval?

Confidence intervals use Z-scores to determine the range within which a population parameter is expected to lie, based on sample statistics.

#20. How are Z-scores used to compare different distributions?

Z-scores standardize different distributions by converting values to a common scale, allowing comparisons between datasets with different means and standard deviations.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

x = np.arange(1, 7)  # Rolling a die (1-6)
pmf = randint.pmf(x, 1, 7)

plt.bar(x, pmf, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

```python
from scipy.stats import bernoulli

p = 0.5  # Probability of success
x = [0, 1]
pdf = bernoulli.pmf(x, p)

print("Bernoulli PDF values:", pdf)
```

4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram

```python
n, p = 10, 0.5
samples = np.random.binomial(n, p, 1000)

plt.hist(samples, bins=10, density=True, alpha=0.7, color='green')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution Histogram')
plt.show()
```

5. Create a Poisson distribution and visualize it using Python

```python
from scipy.stats import poisson

mu = 3  # Mean number of events
x = np.arange(0, 10)
pmf = poisson.pmf(x, mu)

plt.bar(x, pmf, color='purple', alpha=0.7)
plt.xlabel('Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import scipy.stats as st

confidence_level = 0.95
sample_data = np.random.normal(100, 15, 30)
mean = np.mean(sample_data)
margin_error = st.t.ppf((1 + confidence_level) / 2, len(sample_data) - 1) * (np.std(sample_data) / np.sqrt(len(sample_data)))

print("Confidence Interval:", (mean - margin_error, mean + margin_error))
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
confidence_interval = st.norm.interval(0.95, loc=np.mean(sample_data), scale=st.sem(sample_data))
print("Confidence Interval:", confidence_interval)
```
